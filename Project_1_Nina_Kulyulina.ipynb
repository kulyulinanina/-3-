```python
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

def get_city(s):
    s = str(s)
    city = s.split(',')[0].strip()
    if city == 'Москва':
        return 'Москва'
    elif city == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

def get_relocation(s):
    s = str(s)
    s = s.lower()
    if 'не готов' in s or 'не готова' in s:
        return False
    elif 'готов' in s or 'готова' in s or 'хочу' in s:
        return True
    else:
        return False

def get_business_trips(s):
    s = str(s)
    s = s.lower()
    if 'не готов' in s or 'не готова' in s:
        return False
    elif 'готов' in s or 'готова' in s:
        return True
    else:
        return False

df['Город'] = df['Город, переезд, командировки'].apply(get_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(get_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(get_business_trips)

print(df['Город'].unique())
print(df['Готовность к переезду'].unique())
print(df['Готовность к командировкам'].unique())
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
```

```python
def one_hot_encode(df, column):
    categories = df[column].str.get_dummies(sep=', ')
    categories.columns = [col.capitalize() for col in categories.columns]
    return categories

employment_ohe = one_hot_encode(df, 'Занятость')
schedule_ohe = one_hot_encode(df, 'График')

df = pd.concat([df, employment_ohe, schedule_ohe], axis=1)
df.columns
```

```python
project_volunteering = len(df[(df['Проектная работа'] == True) & (df['Волонтерство'] == True)])
print(f"Количество людей, ищущих проектную работу и волонтёрство: {project_volunteering}")

shift_flexible = len(df[(df['Вахтовый метод'] == True) & (df['Гибкий график'] == True)])
print(f"Количество людей, желающих работать вахтовым методом и с гибким графиком: {shift_flexible}")

df.drop(['Занятость', 'График'], axis=1, inplace=True)
df.columns
```

```python
file_path = '/content/drive/My Drive/ExchangeRates.csv'
with open(file_path, 'r') as file:
    currency_df = pd.read_csv(file_path, sep=',')

currency_df.drop(['per', 'time', 'vol'], axis=1, inplace=True)
currency_df.head()
```

```python
# Шаг 1: Преобразование даты в формате datetime
data = df
data['Обновление резюме'] = pd.to_datetime(df['Обновление резюме']).dt.date
currency_df['date'] = pd.to_datetime(currency_df['date']).dt.date

# Шаг 2.1: Извлечение суммы и валюты из столбца "ЗП"
data[['ЗП_сумма', 'ЗП_валюта']] = df['ЗП'].str.split(expand=True)
data['ЗП_сумма'] = data['ЗП_сумма'].astype(float)

# Шаг 2.2: Сопоставление валюты с ISO из справочной таблицы
currency_mapping = {
    'грн.': 'UAH', 'USD': 'USD', 'EUR': 'EUR', 'бел.руб.': 'BYN',
    'KGS': 'KGS', 'сум': 'UZS', 'AZN': 'AZN', 'KZT': 'KZT', 'руб.' : 'RUB'
}
data['ЗП_валюта_ISO'] = data['ЗП_валюта'].map(currency_mapping)

# Шаг 3.1: Присоединение курса валют по дате и валюте
data = pd.merge(
    data,
    currency_df,
    left_on=['ЗП_валюта_ISO', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)

# Шаг 3.2. Заполнение курса рубля для случаев, когда валюта уже в рублях
data['close'] = data['close'].fillna(1)
data['proportion'] = data['proportion'].fillna(1)

# Шаг 4. Расчет зарплаты в рублях
data['ЗП (руб)'] = data.apply(
    lambda row: row['ЗП_сумма'] if row['ЗП_валюта'] == 'RUB'
    else (row['ЗП_сумма'] * row['close']) / row['proportion'], axis=1
)

# Очистка ненужных столбцов
data = data.drop(columns=['ЗП', 'ЗП_сумма', 'ЗП_валюта', 'ЗП_валюта_ISO', 'currency', 'date', 'close', 'proportion'])


# Расчет медианной зарплаты
median_salary = data['ЗП (руб)'].median()
median_salary_thousands = int(round(median_salary / 1000))
print(f"Желаемая медианная заработная плата: {median_salary_thousands} тысяч рублей")
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
data['log_age'] = np.log1p(data['Возраст'])

plt.figure(figsize=(10, 6))
histplot = sns.histplot(data['log_age'], kde=True)

log_age_mean = data['log_age'].mean()
log_age_std = data['log_age'].std()

histplot.axvline(log_age_mean, color='k', lw=2, label='Mean')
histplot.axvline(log_age_mean - 3*log_age_std, color='r', ls='--', lw=2, label='-3 std')
histplot.axvline(log_age_mean + 4*log_age_std, color='g', ls='--', lw=2, label='+4 std')

plt.xlabel('Logarithm of Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age (Logarithmic Scale)')
plt.legend()
plt.text(0.5, 0.9, "Logarithmic distribution is right-skewed", transform=plt.gca().transAxes)
plt.show()
```

```python
data['zscore_log_age'] = (data['log_age'] - log_age_mean) / log_age_std
outliers = data[(data['zscore_log_age'] < -3) | (data['zscore_log_age'] > 4)]

print("Outliers based on logarithmic age:")
print(outliers[['Возраст', 'log_age', 'zscore_log_age']])

data = data[(data['zscore_log_age'] >= -3) & (data['zscore_log_age'] <= 4)]
data.drop(['log_age', 'zscore_log_age'], axis=1, inplace=True)
```
